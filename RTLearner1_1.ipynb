```python
""""""
"""
A simple wrapper for linear regression.  (c) 2015 Tucker Balch

Copyright 2018, Georgia Institute of Technology (Georgia Tech)
Atlanta, Georgia 30332
All Rights Reserved

Template code for CS 4646/7646

Georgia Tech asserts copyright ownership of this template and all derivative
works, including solutions to the projects assigned in this course. Students
and other users of this template code are advised not to share it with others
or to make it available on publicly viewable websites including repositories
such as github and gitlab.  This copyright statement should not be removed
or edited.

We do grant permission to share solutions privately with non-students such
as potential employers. However, sharing with other current or future
students of CS 7646 is prohibited and subject to being investigated as a
GT honor code violation.

-----do not edit anything above this line---
"""

"""
Student Name: Yuanzhang Zheng (replace with your name)
GT User ID: yzheng438   (replace with your User ID)
GT ID: 903649243 (replace with your GT ID)
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

class RTLearner(object):
    """
    This is a Linear Regression Learner. It is implemented correctly.

    :param verbose: If “verbose” is True, your code can print out information for debugging.
        If verbose = False your code should not generate ANY output. When we test your code, verbose will be False.
    :type verbose: bool
    """
    def __init__(self, leaf_size = 1, verbose=False):
        """
        Constructor method
        """
        self.leaf_size = leaf_size
        self.verbose = verbose
        self.tree = None

    def author(self):
        """
        :return: The GT username of the student
        :rtype: str
        """
        return "yzheng438"  # replace tb34 with your Georgia Tech username

    def study_group(self):
      return "yzheng438"

    def create_dataset(self):
      dataset = pd.read_csv('Data/lstanbul.csv').values
      use_dataset = dataset[1:]

      return use_dataset




    def add_evidence(self, data_x, data_y):
        """
        Add training data to learner
        :param data_x: A set of feature values used to train the learner
        :type data_x: numpy.ndarray
        :param data_y: The value we are attempting to predict given the X data
        :type data_y: numpy.ndarray
        """
        self.tree = self.build_tree(data_x, data_y)

    def build_tree(self, data_x, data_y):
    # 如果只有一个样本
    if data_x.shape[0] == 1:
        return np.array([[-1, np.mean(data_y), np.nan, np.nan]])

    # 如果所有目标值都相同
    if np.all(data_y == data_y[0]):
        return np.array([[-1, np.mean(data_y), np.nan, np.nan]])

    # 如果样本数小于等于叶子节点大小或所有目标值相同
    if data_x.shape[0] <= self.leaf_size:
        return np.array([[-1, np.mean(data_y), np.nan, np.nan]])

    # 随机选择一个特征进行分裂
    random_feature = np.random.randint(data_x.shape[1])
    # 选择分裂点
    split_val = np.median(data_x[:, random_feature])

    # 分割数据集
    left_index = data_x[:, random_feature] <= split_val
    right_index = data_x[:, random_feature] > split_val

    # 如果分裂无效
    if np.all(left_index) or np.all(right_index):
        return np.array([[-1, np.mean(data_y), np.nan, np.nan]])

    # 构建左子树和右子树
    left_tree = self.build_tree(data_x[left_index], data_y[left_index])
    right_tree = self.build_tree(data_x[right_index], data_y[right_index])

    # 构建当前节点
    root = np.array([[random_feature, split_val, 1, left_tree.shape[0] + 1]])

    return np.vstack((root, left_tree, right_tree))



    def select_random_feature(self, data_x, data_y):
      correlations = []
      numbers_features = data_x.shape[1]

      random_feature_index = np.argmax.randint(numbers_features)

      return random_feature_index

    def select_split_val(self, data_x, random_feature_index):
        random_feature_values = data_x[:, random_feature_index]
        split_val = np.median(random_feature_values)

        return split_val




    def query(self, points):
        """
        Estimate a set of test points given the model we built.

        :param points: A numpy array with each row corresponding to a specific query.
        :type points: numpy.ndarray
        :return: The predicted result of the input data according to the trained model
        :rtype: numpy.ndarray
        """
        return np.array([self.query_point(point) for point in points])


    def query_point(self, point):
        node = 0
        # 当当前节点不是叶节点时，继续遍历树
        while self.tree[node, 0] != -1:
            # 判断点在分裂点的左边还是右边
            if point[int(self.tree[node, 0])] <= self.tree[node, 1]:
                # 移动到左子树
                node += int(self.tree[node, 2])
            else:
                # 移动到右子树
                node += int(self.tree[node, 3])
        # 返回叶节点的值
        return self.tree[node, 1]






if __name__ == "__main__":
    print("the secret clue is 'zzyzx'")
```
