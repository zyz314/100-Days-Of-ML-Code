```python
# 导入必要的库
# 导入 pandas 作为 pd
# 导入 numpy 作为 np
# 导入 matplotlib.pyplot 作为 plt
# 读取数据的辅助函数
# 从 util 导入 get_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import util
from util inport get_data



# 定义技术指标函数

def MACD(df, fast_span, slow_span, signal_span):
    # 输入：股票数据 DataFrame，快速EMA时间跨度 fast_span，慢速EMA时间跨度 slow_span，信号线时间跨度 signal_span
    # 计算：移动平均线收敛背离
    # ema_fast = 计算 df 中 '价格' 列的快速指数加权移动平均，时间跨度为 fast_span
    ema_fast = df['price'].ewm(span=fast_span, adjust=False).mean()
    # ema_slow = 计算 df 中 '价格' 列的慢速指数加权移动平均，时间跨度为 slow_span
    ema_slow = df['price'].ewm(span = slow, adjust = False).mean()
    macd = ema_fast - ema_slow
    # signal = 计算 macd 的指数加权移动平均，时间跨度为 signal_span
    signal = macd.ewm(span = signal_span, adjust = False).mean()\

    # 生成图表：# 绘制MACD图表
    plt.figure(figsize = (12.9))
    plt.title("Indicator by using MACD")
    # 画图，x轴为 df 中的 '日期' 列，y轴为 macd，标注为 'MACD'
    plt.plot(df.index, macd, label = 'MACD', color = 'blue')
    plt.plot(df.index, signal, label = "signal", color = 'red')
    # 画图，x轴为 df 中的 '日期' 列，y轴为 signal，标注为 '信号线'
    plt.xlabel("Date")
    plt.ylabel('Value')
    plt.legend(loc="best")
    plt.grid(True)

    # 添加图例
    # 保存图表，文件名为 'MACD.png'
    plt.savefig('MACD.png')
    # 返回：macd 线和信号线 (macd, signal)

    return macd, signal





def EMA(df, span):
    # 输入：股票数据 DataFrame，时间跨度 span

    ema = df['price'].ewm(span = span, adjust = False).mean()
    # 计算：指数移动平均线
    # ema = 计算 df 中 '价格' 列的指数加权移动平均，时间跨度为 span
    # 生成图表：
    plt.figure(figsize = (12.9))
    plt.plot(df.index, df['price'], label = 'price', color = 'blue')
    plt.plot(df.index, ema, label = 'EMA', color = 'red' )
    plt.xlabel("data")
    plt.ylabel("price")
    plt.legend(loc = 'best')
    plt.grid(True)
    plt.savefig("EMA.png")
    plt.close

    return ema


def RSI(df, window):
    # 输入：股票数据 DataFrame，窗口期 window
    # 计算：相对强弱指数
    # delta = 计算 df 中 '价格' 列的变化
    delta = df["price"].diff()
    # gain = 计算 delta 中正值的滚动平均，窗口期为 window
    # loss = 计算 delta 中负值的滚动平均，窗口期为 window
    gain =delta.where(delta>0, 0).rolling(window = window).mean()
    loss = delta.where(delta<0, 0).rolling(window = window).mean()


    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    # 生成图表：
    plt.figure(figsize = (12, 9))
    plt.title("RSI")
    plt.plot(df.index, rsi, label = 'RSI', color = 'red')
    plt.axhline(y = 70, color = 'blue', linestyle = '--')
    plt.axhline(y = 30, color = 'orange', linestyle = '--')
    plt.xlabel("Date")
    plt.ylabel("RSI")
    plt.legend(loc = 'best')
    plt.grid(True)
    plt.savefig("RSI.png")
    plt.close

    return rsi



    # 画图，x轴为 df 中的 '日期' 列，y轴为 rsi，标注为 'RSI'
    # 添加图例
    # 保存图表，文件名为 'RSI.png'
    # 返回：相对强弱指数 rsi

def SMA(df, window):
    # 输入：股票数据 DataFrame，窗口期 window
    # 计算：简单移动平均线
    sma = df['price'].rolling(window= window).mean()


    # sma = 计算 df 中 '价格' 列的滚动平均，窗口期为 window
    # 生成图表：
    plt.figure(figsize = (12,9))
    plt.plot(df.index, df['price'], label = 'price', color = 'red')
    plt.plot(df.index, sam, lable = 'SMA', color = 'blue')
    plt.title('SMA')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend(loc= 'best')
    plt.grid(True)
    plt.savefig('SMA.png')
    plt.close()


def BollingerBands(df, window):
    # 输入：股票数据 DataFrame，窗口期 window
    # 计算：布林带
    # sma = 计算 df 中 '价格' 列的滚动平均，窗口期为 window
    sam = df['price'].rolling(window = window).mean()

    std = df['price'].rolling(window = window).std()

    upper_band = sma + (2 * std)
    lower_band = sma - (2 * std)
    plt.figure(figsize = (12, 9))
    plt.plot(df.index, df['price'], label = 'price', color = 'red')
    plt.plot(df.index, sma, label = 'middle track', color = 'blue')
    plt.plot(df.index, upper_band, label = 'upper track', color = 'green')
    plt.plot(df.index, lower_band, label = ' lower track', color = 'orange')
    plt.fill_between(df.index, upper_band, lower_band, color = "light gray", alpha = 0.25)
    plt.title("Bollinger Bands")
    plt.xlabel('Date')
    plt.ylabel("Price")
    plt.legend(loc = 'best')
    plt.grid(True)
    plt.savefig('Bollinger_Bands.png')
    plt.close()

    return lower_band, upper_band


# 主函数用于运行所有指标和生成图表
# 函数 run_all_indicators(df):
    # 运行所有技术指标函数
    # sma = 调用 SMA(df, window=20)
    # upper_band, lower_band = 调用 BollingerBands(df, window=20)
    # ema = 调用 EMA(df, span=20)
    # rsi = 调用 RSI(df, window=14)
    # macd, signal = 调用 MACD(df, fast_span=12, slow_span=26, signal_span=9)
    # 返回：所有指标的结果 (sma, (upper_band, lower_band), ema, rsi, (macd, signal))

# 调用主函数运行指标
if __name__ == "__main__":
    # df = 调用 get_data(['JPM'], dates=pd.date_range('2008-01-01', '2009-12-31'))
    # sma, bollinger_bands, ema, rsi, macd = 调用 run_all_indicators(df)
```
